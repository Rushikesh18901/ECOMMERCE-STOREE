# MongoDB Experiments - From Scratch

This notebook demonstrates basic MongoDB operations using pymongo.
Each cell shows a core concept of MongoDB operations.

```python
# 🧪 🔹 CELL 1 — Install MongoDB Driver
# Run this only once
!pip install pymongo
```

```python
# 🧪 🔹 CELL 2 — Import Library
from pymongo import MongoClient
```

```python
# 🧪 🔹 CELL 3 — Connect to MongoDB
# Local MongoDB connection
client = MongoClient("mongodb://localhost:27017/")

# Create / connect database
db = client["ecommerce_db"]

# Create / connect collection
collection = db["products"]

print("Connected to MongoDB ✅")
```

```python
# 🧪 🔹 CELL 4 — Insert One Product
product = {
    "name": "Men Casual Shirt",
    "price": 999,
    "category": "men"
}

collection.insert_one(product)

print("1 Product Inserted ✅")
```

```python
# 🧪 🔹 CELL 5 — Insert Multiple Products
products = [
    {"name": "Jeans", "price": 1499, "category": "men"},
    {"name": "Kurti", "price": 999, "category": "women"},
    {"name": "Kids Jacket", "price": 1299, "category": "kids"}
]

collection.insert_many(products)

print("Multiple Products Inserted ✅")
```

```python
# 🧪 🔹 CELL 6 — View All Products
all_products = collection.find()

for p in all_products:
    print(p)
```

```python
# 🧪 🔹 CELL 7 — Find One Product
product = collection.find_one({"name": "Jeans"})

print(product)
```

```python
# 🧪 🔹 CELL 8 — Filter Products
men_products = collection.find({"category": "men"})

for p in men_products:
    print(p)
```

```python
# 🧪 🔹 CELL 9 — Update Product
collection.update_one(
    {"name": "Jeans"},
    {"$set": {"price": 1399}}
)

print("Product Updated ✅")
```

```python
# 🧪 🔹 CELL 10 — Delete One Product
collection.delete_one({"name": "Kids Jacket"})

print("Product Deleted ✅")
```

```python
# 🧪 🔹 CELL 11 — Delete All Products
collection.delete_many({})

print("All Products Deleted ✅")
```

```python
# 🧪 🔹 CELL 12 — Count Documents
count = collection.count_documents({})

print("Total Products:", count)
```

```python
# 🧪 🔹 CELL 13 — Drop Collection
collection.drop()

print("Collection Dropped ✅")
```
